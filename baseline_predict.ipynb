```python
davis_mse = []
davis_ci = []
davis_r2m = []

kiba_mse = []
kiba_ci = []
kiba_r2m = []

model_name = []

with open("./baseline_predict.csv",'r') as f:
    lines = f.readlines()
    for l in lines[1:9]:
        ls = l.split(",")
        davis_mse.append(ls[3])
        davis_ci.append(ls[2])
        davis_r2m.append(ls[4])
        model_name.append(ls[1])
    
    for l in lines[9:]:
        ls = l.split(",")
        kiba_mse.append(ls[3])
        kiba_ci.append(ls[2])
        kiba_r2m.append(ls[4])
```

```python
import matplotlib.pyplot as plt# 导入模块
import numpy as np

plt.figure()
fig, ax = plt.subplots()
plt.xticks(rotation=45)
ax.plot(model_name, [float(i) for i in davis_mse[:]])
plt.xlabel("Model", fontdict={'size': 18})
plt.ylabel("MSE", fontdict={'size': 18})
plt.title("Davis MSE", fontdict={'size': 20})
# plt.savefig("PoolRate.png")

plt.figure()
fig, ax = plt.subplots()
plt.xticks(rotation=45)
ax.plot(model_name, [float(i) for i in davis_ci[:]])
plt.xlabel("Model", fontdict={'size': 18})
plt.ylabel("CI", fontdict={'size': 18})
plt.title("Davis CI", fontdict={'size': 20})
# plt.savefig("PoolRate.png")

plt.figure()
fig, ax = plt.subplots()
plt.xticks(rotation=45)
ax.plot(model_name, [float(i) for i in davis_r2m[:]])
plt.xlabel("Model", fontdict={'size': 18})
plt.ylabel("R2M", fontdict={'size': 18})
plt.title("Davis R2M", fontdict={'size': 20})
# plt.savefig("PoolRate.png")

plt.show()
```
