```python
mse = []
ci = []

with open("./hyperparameters_predict.csv",'r') as f:
    lines = f.readlines()[2:]
    for l in lines:
        ls = l.split(",")
        mse.append(ls[3])
        ci.append(ls[-2])
        
# print(mse)
# print(ci)

x = [i*0.1 for i in range(10)][1:]
```
