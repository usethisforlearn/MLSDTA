```python
import matplotlib.pyplot as plt# 导入模块
import numpy as np


davis_r = []
davis_p = []
with open("./valuedavis.txt") as f1:
    lines = f1.readlines()
    lines = lines[1:]
    for line in lines:
        g,p = line.split(",")
        davis_r.append(float(g))
        davis_p.append(float(p))
        
kiba_r = []
kiba_p = []
with open("./valuekiba.txt") as f1:
    lines = f1.readlines()
    lines = lines[1:]
    for line in lines:
        q,a = line.split(",")
        if(float(q)>=9):
            kiba_r.append(float(q))
            kiba_p.append(float(a))
        
print(len(davis_r))
print(len(kiba_r))



plt.figure()

plt.scatter(davis_r, davis_p, c="b",  s=1, alpha=0.7)

plt.xticks(range(5, 12))
plt.yticks(range(5, 12))

plt.xlim(4.5, 11)
plt.ylim(4.5, 11)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.gca().set_aspect('equal')

x_line = np.linspace(5, 11)
plt.plot(x_line, x_line, 'k--')

plt.xlabel("Actual Value", fontdict={'size': 18})
plt.ylabel("Predicted Value", fontdict={'size': 18})
# plt.title("Davis", fontdict={'size': 20})


# plt.savefig("Davisfig.png")
plt.show()
```
