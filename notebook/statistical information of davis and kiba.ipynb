# Davis的药物

```python
import json
import tqdm
import matplotlib.pyplot as plt# 导入模块
import numpy as np
import seaborn as sns

length = []
with open('./davis_ligands_can.txt','r') as f:
    lines = f.readlines()[0]
    davis_drug = json.loads(lines)
    for i in list(davis_drug.values()):
        length.append(len(i))
# print(length)



# plt.figure() #初始化一张图
# x = length
# plt.hist(x,bins=25,edgecolor='black')  #直方图关键操作 ,normed=True
# plt.grid(alpha=0.5,linestyle='-.') #网格线，更好看 
# plt.xlabel('Length of ligand SMILES')  
# plt.ylabel('Number of ligands')  
# plt.title(r'Davis') 
# plt.savefig("Davis_ligands.png")
# plt.show()

sns.set(style="darkgrid")
sns.distplot(length,hist=True, kde=True,bins=15)
plt.xlabel('Length of ligand SMILES')  
plt.ylabel('Density of ligands')  
plt.title(r'Davis') 
plt.savefig("1Davis_ligands.png")
plt.show()
```
